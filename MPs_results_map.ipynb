```python
import pandas as pd
from google.colab import files
uploaded = files.upload()
uploaded = files.upload()

DataFrame1 = pd.read_csv('MPs election results.csv')
DataFrame2 = pd.read_csv('Parliament_con.csv')
```

```python
combined_table = pd.merge(DataFrame1, DataFrame2, on='ons_id', how='inner')
print(combined_table.head())
```

```python
combined_table.to_csv("combined_table.csv", index=False)
```

```python
import folium
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

parties = combined_table['party_name'].unique()
colors = plt.cm.tab20.colors[:len(parties)]
party_colors = {party: f"rgb({int(r*255)}, {int(g*255)}, {int(b*255)})" for party, (r, g, b) in zip(parties, colors)}
```

```python
map = folium.Map(location=[combined_table['center_lat'].mean(), combined_table['center_lon'].mean()], zoom_start=6)

for _, row in combined_table.iterrows():
    folium.CircleMarker(
        location=[row['center_lat'], row['center_lon']],
        radius=2,
        color=party_colors[row['party_name']],
        fill=True,
        fill_color=party_colors[row['party_name']],
        fill_opacity=1,
        tooltip=f"Constituency: {row['constituency_name']}<br>Party: {row['party_name']}"
    ).add_to(map)

legend_html = """
<div style="position: fixed;
            bottom: 20px; left: 20px; width: 200px; height: auto;
            background-color: white; z-index:9999; font-size:6px;
            border:2px solid grey; border-radius:5px; padding:10px;">
    <b>Party Legend</b><br>
    {}
</div>
""".format("<br>".join([f'<span style="display: inline-block; width: 10px; height: 10px; background-color: {color};"></span> {party}' for party, color in party_colors.items()]))

map.get_root().html.add_child(folium.Element(legend_html))

map
```
